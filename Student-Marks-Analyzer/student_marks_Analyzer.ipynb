```python
marks={
    "Klaus":[90,86,77],
    "caroline":[80,79,88],
    "Demon":[96,70,80],
    "Elena":[99,67,70]
}
```

```python
print("Student wise Averages")
for name in marks:
    avg=sum(marks[name])/3
    print(name,"-> Average : ",round(avg,2))    #This keeps 2 decimal place 
```

```python
print("Topper of Class")
max_total=0 #variable stores the highest total marks found so far
topper="" #variable stores the name of the topper
for name in marks:
    total=sum(marks[name]) #marks[name] gives the list of 3 subject marks for the student
                           #sum() adds all those marks
    if total>max_total:
        max_total=total   #Update max_total to the new higher value
        topper=name
print("Topper : ",topper, "with" ,max_total ,"marks")
```

```python
print("Subject wise Average")
student_count=len(marks)

sub1=0
sub2=0
sub3=0

for name in marks:
    sub1=sub1+marks[name][0]
    sub2=sub2+marks[name][1]
    sub3=sub3+marks[name][2]

print("Subject1 : ",round(sub1/student_count,2))
print("Subject2 : ",round(sub2/student_count,2))
print("Subject3 : ",round(sub3/student_count,2))
```
