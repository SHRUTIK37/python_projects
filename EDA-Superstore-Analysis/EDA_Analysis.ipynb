## 1. Bar Chats

```python
df["Category"].value_counts().plot(kind="bar")
plt.title("Count by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
df['Region'].value_counts().plot(kind='bar', color='purple')
plt.title("Count by Region")
plt.xlabel('Region')
plt.ylabel('Count')
plt.show()
```

```python
df["Segment"].value_counts().plot(kind="bar")
plt.title("Count by Segment")
plt.xlabel("Segment")
plt.ylabel("Count")
plt.show()
```

```python
df["State"].value_counts().plot(kind="bar")
plt.title("Count by stat")
plt.xlabel("State")
plt.ylabel("Count")
plt.show()
```

```python
df["Sub-Category"].value_counts().plot(kind="bar", color='c')
plt.title("Count by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Count")
plt.show()
```

## 2. Scatter Plot

```python
plt.scatter(df['Sales'], df["Profit"], color='pink')
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
plt.scatter(df["Discount"], df["Profit"])
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

## 3. Histogram

```python
plt.hist(df['Sales'], bins=30)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df["Profit"], bins=30, color='pink')
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["Quantity"], bins=20, color='green')
plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df["Discount"], bins=20, color='blue')
plt.title("Discount Distribution")
plt.xlabel("Discount")
plt.ylabel("Frequency")
plt.show()
```

##4. Heatmap

```python
numeric_df = df[["Sales", "Profit", "Discount", "Quantity"]]
```

```python
corr=numeric_df.corr()
```

```python
sns.heatmap(corr,annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.heatmap(corr,annot=True, cmap='bwr') #coolwarm, viridis,seismic
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.heatmap(corr,annot=True, cmap='summer')
plt.title('Correlation Heatmap')
plt.show()
```
