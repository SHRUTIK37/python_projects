```python
account = {}
```

```python
def create_account():
    name = input("Enter name: ").strip()
    balance = int(input("Enter initial balance: "))
    account["name"] = name
    account["balance"] = balance
    print("Account created!")
```

```python
def check_balance():
    if account:
        print(account["name"], "balance:", account["balance"])
    else:
        print("No account found. Please create an account first.")
```

```python
def deposit():
    if account:
        amt = int(input("Enter amount to deposit: "))
        account["balance"] += amt
        print("Deposit successful!")
    else:
        print("No account found. Please create an account first.")
```

```python
def withdraw():
    if account:
        amt = int(input("Enter amount to withdraw: "))
        if amt <= account["balance"]:
            account["balance"] -= amt
            print("Withdrawal successful!")
        else:
            print("Insufficient balance!")
    else:
        print("No account found. Please create your account first.")
```

```python
while True:
    print("\nWelcome to MyBank!")
    print("1. Create Account")
    print("2. Check Balance")
    print("3. Deposit")
    print("4. Withdraw")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        create_account()
    elif choice == "2":
        check_balance()
    elif choice == "3":
        deposit()
    elif choice == "4":
        withdraw()
    elif choice == "5":
        print("Thank you ")
        break
    else:
        print("Invalid choice.Enter your choice from above")
```
